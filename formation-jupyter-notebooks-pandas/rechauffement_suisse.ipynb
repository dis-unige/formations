## Comparaison avec l'ére préindustrielle

Selon Meteosuisse l'ère préindustrielle est comprise entre 1850 et 1900

```python
meteo
```

```python
meteo.loc[1968]
```

```python
meteo_preind = meteo.loc[:1900]
meteo_preind
```

```python
# moyenne
temp_moyenne_preind = meteo_preind['Temperature'].mean()
temp_moyenne_preind
```

```python
# moyenne
precipitation_moyenne_preind = meteo_preind['Precipitation'].mean()
precipitation_moyenne_preind
```

```python
# graphique des moyennes
meteo_preind['Temperature'].plot()
```

```python
# graphique des moyennes
meteo_preind['Temperature'].groupby(by='Year').mean().plot()
```

```python
# graphique des moyennes
meteo_preind['Temperature'].groupby(by='Year').mean().plot(kind='hist')
```

```python
# graphique des moyennes
meteo_preind['Temperature'].groupby(by='Year').mean().plot(kind='density')
```

## Ecarts par rapport à la température moyenne préindustrielle

```python
meteo
```

```python
meteo_post = meteo.loc[1901:2023].groupby(by='Year').mean()
meteo_post
```

```python
meteo_post['Temperature_ecart'] = meteo_post['Temperature'] - temp_moyenne_preind
meteo_post
```

```python
meteo_post['Précipitation_ecart'] = meteo_post['Precipitation'] - precipitation_moyenne_preind
meteo_post
```

```python
meteo_post['Temperature_ecart'].plot()
```

```python
meteo_post['Précipitation_ecart'].plot()
```
