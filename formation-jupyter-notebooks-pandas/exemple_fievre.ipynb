# Exemples avec Jupyter Notebook et Pandas

Cas déclarés de fièvre typhoïde, par canton, de 1894 à 1958

Source : https://hsso.ch/fr/2012/d/29

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = ('https://hsso.ch/get/D.29.xlsx')
```

```python
df = pd.read_excel(url, skiprows=2, nrows=66)
```

```python
df.head(10)
```

```python
# supprimer la première ligne qui est vide
df = df.iloc[1:]
df
```

```python
# supprimer la dernière colonne qui est répétée
del df['Année']
df
```

```python
# indexer par année
df = df.set_index('Jahr')
df
```

```python
# Analyse rapide de données numériques
df.describe()
```

```python
# graphique pour toute la suisse 
df['average'] = df.mean(numeric_only=True, axis=1)
df[['CH', 'average']].plot(figsize=(15, 6), label='CH toal', lw=2)
```

```python
# graphique pour les canton de Genève, Zurich, Berne et Vaud
df[['GE', 'ZH', 'BE', 'VD']].plot(figsize=(15, 6))
```

```python
from ipywidgets import interact

@interact
def plot(n=(1, 50)):
    fig, ax = plt.subplots(1, 1, figsize=(15, 10))
    df[['GE', 'ZH', 'BE', 'VD']].rolling(window=n).mean().plot(ax=ax)
    ax.set_ylim(0, 200)
    ax.legend(loc='upper right', fontsize='small', bbox_to_anchor=(1.1, 1))
    plt.show()
```
